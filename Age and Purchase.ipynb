```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
Data = pd.read_csv("shopping_trends_updated.csv")
ar_data = np.array(Data)
```

```python
#Create an Array of Age and Purchase Amount
arrayOfAge = ar_data[:,1].astype(int)

arrayOfPurchase = ar_data[:,5].astype(int)
```

```python
#Assigning Variables for X and Y
x=arrayOfAge;
y=arrayOfPurchase;

#Creation of the Scatter Plot
sns.set(style="whitegrid")
plt.figure(figsize=(15,15))
sns.scatterplot(x=arrayOfAge, y=arrayOfPurchase)
plt.title('Scatter Plot of Age vs. Purchase Amount')
plt.xlabel('Age')
plt.ylabel('Purchase Amount')


slope, intercept, r, p, std_err = stats.linregress(x, y) # Calculating the linear regression parameters
def myfunc(x):
  return slope * x + intercept # Defining a linear function based on the regression parameters
mymodel = list(map(myfunc, x)) # Generating predicted values using the linear function

print(r)

plt.plot(x, mymodel)
plt.show()
```
