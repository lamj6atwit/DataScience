```python
#Imports required to run the program
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
Data = pd.read_csv("shopping_trends_updated.csv")
ar_data = np.array(Data)
```

```python
#Create an Array of Reviews and Purchase Amount
arrayOfPurchaseAmount = ar_data[:,5].astype(int)

arrayOfReviews = ar_data[:,10].astype(float)
```

```python
#Assigning Variables for X and Y
x=arrayOfReviews;
y=arrayOfPurchaseAmount;

#Creation of the Scatter Plot
sns.set(style="whitegrid")
plt.figure(figsize=(15,15))
sns.scatterplot(x=arrayOfReviews, y=arrayOfPurchaseAmount)
plt.title('Scatter Plot of Purchase Amount VS. Reviews')
plt.xlabel('Reviews')
plt.ylabel('Purchase Amount')


slope, intercept, r, p, std_err = stats.linregress(x, y) # Calculating the linear regression parameters
def myfunc(x):
  return slope * x + intercept # Defining a linear function based on the regression parameters
mymodel = list(map(myfunc, x)) # Generating predicted values using the linear function

print(r)

plt.plot(x, mymodel)
plt.show()

```
